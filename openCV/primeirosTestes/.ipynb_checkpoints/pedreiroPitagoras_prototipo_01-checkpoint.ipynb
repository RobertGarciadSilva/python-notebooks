### Primeiro teste PedreiroPitagoras (testes_openCV)

O objetivo é tentar estimar reais comprimentos, medidos por uma regua, por exemplo, em uma imagem em que um comprimento já é fornecido, ou seja, utilizando um comprimento já medido de uma imagem, tentar estimar outros comprimentos.

```python

import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python

img01 = cv2.imread('./img/garrafa01.jpeg')
```

```python
imgAltura, imgLargura = img01.shape[0], img01.shape[1]
```

```python

fig, axe = plt.subplots(figsize=(12,8))
axe.imshow(cv2.cvtColor(img01, cv2.COLOR_BGR2RGB)) # "as opencv loads in BGR format by default, we want to show it in RGB."
#plt.imshow(img01)
plt.show()
```

```python

#Achando alguns pontos de interesse na imagem usando vlines

fig, axe = plt.subplots(figsize=(18,10))
axe.imshow(cv2.cvtColor(img01, cv2.COLOR_BGR2RGB)) # "as opencv loads in BGR format by default, we want to show it in RGB."
#plt.imshow(img01)
axe.vlines(445,0,imgAltura-1, color='blue') # O less one deve-se ao fato dos vetores começarem do 0, logo uma imagem com 1280 pixels de altura, irá ser armazenada de 0 á 1279
axe.vlines(550,0, imgAltura-1, color='blue')
axe.hlines(380,0, imgLargura-1, color='blue')

plt.show()

pontoA = [445, 380] # coordenada (x,y)
pontoB = [550, 380] # coordenada (x,y)

numPixelsEntreAB = pontoB[0] - pontoA[0] # pixels
numPixelsEntreAB_2 = np.sqrt((pontoB[0]-pontoA[0])**2 + (pontoB[1]-pontoA[1])**2); #print(numPixelsEntreAB_2) # mais geral
comprimentoEmpiricoAB = 3.3 # 3cm, real valor medido, utilizado como referência empirica paras as outras estimativas.
pixelsPorComprimento = numPixelsEntreAB/comprimentoEmpiricoAB # pixels por centimetro (número de pixels por centimetro)

print('Por experiencia a distância, comprimento, (medindo na régua) entre o ponto A e B = {0} cm'.format(comprimentoEmpiricoAB), end='\n')
print('O número de pixels entre A e B = {0} pixels'.format(numPixelsEntreAB))
print('Quantidade de pixels por centimetro = {0:.3} pixels/cm'.format(pixelsPorComprimento))
```

```python

fig, axe = plt.subplots(figsize=(12,8))
axe.imshow(cv2.cvtColor(img01, cv2.COLOR_BGR2RGB)) # "as opencv loads in BGR format by default, we want to show it in RGB."
axe.hlines(355,0,imgLargura-1, color='blue')
axe.hlines(980,0,imgLargura-1, color='blue')
axe.vlines(500,0, imgAltura-1, color='blue')
plt.show()

pontoC = [500,355]
pontoD = [500,980]

numPixelsEntreCD = pontoD[1] - pontoC[1]
comprimentoEstimadoCD = numPixelsEntreCD/pixelsPorComprimento;

print('O número de pixels entre C e D = {0} pixels'.format(numPixelsEntreCD))
print('Utilizando o número de pixels/cm acima {0:.3}, temos que uma estimativa para o comprimento real entre C e D = {1:.3} cm'.format(pixelsPorComprimento,comprimentoEstimadoCD))
print('O valor real, medido com auxílio de uma reguá para o comprimento de CD é próximo de {0} cm'.format(20))
```

```python
fig, axe = plt.subplots(figsize=(20,12))
axe.imshow(cv2.cvtColor(img01, cv2.COLOR_BGR2RGB)) # "as opencv loads in BGR format by default, we want to show it in RGB."
axe.vlines(815,0, imgAltura-1, color='red')
axe.vlines(880,0, imgAltura-1, color='red')
axe.hlines(840,0, imgLargura-1, color='red')
plt.show()

pontoE = [815,840]
pontoJ = [880,840]

numPixelsEntreEJ = pontoJ[0]-pontoE[0]
comprimentoEstimadoEJ = numPixelsEntreEJ/pixelsPorComprimento



print('O número de pixels entre E e J = {0} pixels'.format(numPixelsEntreEJ))
print('Utilizando o número de pixels/cm acima {0:.3}, temos que uma estimativa para o comprimento real entre E e J = {1:.3} cm'.format(pixelsPorComprimento,comprimentoEstimadoEJ))
print('O valor real, medido com auxílio de uma reguá para o comprimento de EJ é próximo de {0} cm'.format(2.7))
```
