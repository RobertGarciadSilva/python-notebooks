## Simple Examples plot vetores

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python

vet = np.array([[1,0,1],[0,3,3]])
origin = np.array([[0,1,0],[0,0,0]])


fig = plt.figure(figsize=(7,7))
plt.gca()
plt.quiver(origin[0], origin[1], vet[0], vet[1], angles='xy', scale_units='xy', scale=1, color=['black', 'black','red'])
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.title('Soma vetorial', size='15')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python

arange_x, arange_y = np.arange(-2,3,1), np.arange(-2,3,1)

X, Y = np.meshgrid(arange_x, arange_y)
U, V = X, Y

fig = plt.figure(figsize=(10,10))
plt.quiver(X,Y, U, V, angles='xy', scale_units='xy', scale=1)
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()
```
