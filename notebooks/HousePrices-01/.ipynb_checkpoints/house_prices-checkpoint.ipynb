```python

dataNumeric = data.select_dtypes(include=['int64','float64'])
```

```python

dataNumeric.info()
```

```python
dataNumeric = dataNumeric.drop(['Id','SalePrice'], axis=1)
```

```python

colsNumeric = dataNumeric.columns.values

fig, ax = plt.subplots(6,6, figsize=(18,18))
fig.subplots_adjust(hspace=0.8, wspace=0.9)

n = 0
for axe in ax.flat:
    col = colsNumeric[n]
    axe.scatter(x=dataNumeric[col], y=data['SalePrice'], alpha=0.6, c='red'); n += 1
    axe.set_title('SalePrice vs '+ col)
    axe.set_xlabel(col)
    axe.set_ylabel('SalePrice')

plt.show()
```
