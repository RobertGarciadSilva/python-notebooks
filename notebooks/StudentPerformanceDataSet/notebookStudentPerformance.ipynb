```python
#Libraries
import numpy as np
import pandas as pd
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
```

```python
#import data

data = pd.read_csv('./data/student/student-mat.csv', sep=';')
```

```python
#some elements
data.head(5)
```

```python
data.info()
```

```python
data.describe()
```

```python
#columns
data.columns.values

columsFeatures = ['G1','G2','G3','failures','studytime','absences'] #Used this features based on reference [1]
columLabel = 'G3'

#selecting data features
data = data[columsFeatures]; data
```

```python

x = np.array(data.drop([columLabel],axis='columns')) # array com as features ['G1','G2','failures','studytime','absences']
y = np.array(data[columLabel]) # array com label [G3]

```

```python
#Applying Linear_model_Regression

xTrain, xTest, yTrain, yTest = sklearn.model_selection.train_test_split(x,y, test_size=0.1)
# Spliting the data in train and test, selecting 10% of data to test and 90% to train.

linear = linear_model.LinearRegression()
linear.fit(xTrain, yTrain)
accuracy = linear.score(xTest,yTest)
```

```python
# Some print out after applying Linear_model_Regression

print("Accuracy: ", accuracy)
print("Some info about model:")
print("Coefficient: \n",linear.coef_)
print("Intercept: \n", linear.intercept_)
```

```python
# Comparing preticted, using de model build above, and the real values

compare = []

predicted = linear.predict(xTest) #Applying de model constructed above in data split to teste, dataTest.

for i in range(0, len(xTest)):
    compare.append(["{0:.3f}".format(predicted[i]), xTest[i],yTest[i]])

# where: 
#  predicted[i], represent the value predicted by model build above, modelLinearRegression.
#  xTest[i], represent the values of features used by model to predicted yTest.
#  yTest[i], represent the real value.
   
    
    
    
compareDataFrame = pd.DataFrame(compare, columns=['Predicted', 'FeaturesTest', 'RealValue/LabelTest'])
compareDataFrame
```
