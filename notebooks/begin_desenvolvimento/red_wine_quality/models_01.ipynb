```python

# Import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; 

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score

import shap
```

### Aplicando, treinando modelo 01: RandomForestRegressor

```python
#Import train data
wine = pd.read_csv('./split/train.csv')
```

```python
wine.head()
```

```python
X = wine.drop(['quality'], axis=1)
y = wine['quality']
```

```python

model = RandomForestRegressor(max_depth=6, random_state=42, n_estimators=10)
model.fit(X, y)
```

### Testando modelo 01: RandomForestRegressor

```python
# Import test data
test = pd.read_csv('./split/test.csv')
```

```python
X_test = test.drop(['quality'], axis=1) #features
y_test = test['quality'] #label
```

```python

y_pred = model.predict(X_test)
model.score(X_test, y_test)
```

### Interpretability

```python
shap_values = shap.TreeExplainer(model).shap_values(X)
shap.summary_plot(shap_values, X)
```
