```python
#Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

```

```python
#Import train data

wine = pd.read_csv('./split/train.csv')
```

O objetivo é prever a variável <b>quality</b>, nossa Label. Abaixo é plotado um heatmap com as correlações da vaviável alvo com todas as outras features.

### Correlation

```python

quality_corr = wine.corr()['quality'].sort_values(ascending=False)
#Obtém a matriz de correlação, pega a coluna de correção de "quality" e ordena pelas features que apresentam maior correlação com "quality".

#PLOT
fig, axe = plt.subplots()
sns.heatmap(quality_corr.values.reshape(12,1), annot=True, yticklabels=quality_corr.index, linewidths=.5)
axe.set_title("Quality correlation")

plt.show()
```


#Correlação de todas as features vs todas features

fig, axe = plt.subplots(figsize=(10,6))

Observa-se que:<br><b>alcohol</b> apresenta a maior correlação positiva e <b>volatile acidity</b> maior correlação negativa, ou seja, quanto maior a quantidade de alcohol maior é a qualidade do Wine e quanto maior volatile acidity menor a qualidade do mesmo.

#### Look to variance features 

```python

fig, axe = plt.subplots(1,2, figsize= (14,6), sharey=True)
sns.histplot(data=wine, y='fixed acidity',ax = axe[0], binwidth=0.2)
sns.boxplot(data=wine, y='fixed acidity', ax= axe[1])
plt.show()
```
