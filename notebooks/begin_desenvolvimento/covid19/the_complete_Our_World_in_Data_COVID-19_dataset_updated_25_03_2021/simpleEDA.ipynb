```python
#libraries
import numpy as np
import pandas as pd
import seaborn as sns; sns.set()

import bar_chart_race as bcr
```

```python
#import data
data = pd.read_csv('./owid-covid-data.csv')
```

```python
u = data[data['iso_code'] == 'BRA']
u = pd.concat([u, data[data['iso_code'] == 'USA']])
```

```python
df = data.pivot(index='date', columns='iso_code', values='total_deaths')
```

```python
df
```

```python
df2 = df.dropna(axis=1, how='all')
```

```python
df2
```

```python
temp = df2.isnull().sum().sort_values(ascending=False); temp
```

```python
temp = df2.isnull().sum().sort_values(ascending=False)
cols_drop = []

for val, key in zip(temp, temp.keys()):
    if(val > 60):
        cols_drop.append(key)

len(cols_drop)
```

```python
df4 = df2.drop(axis=1, columns=cols_drop)
```

```python
df4 = pd.concat([df4, df2['BRA']], axis=1)
```

```python
df4 = df4.dropna()
```

```python
df4
```

```python
df5 = df4[['BRA','USA','FRA','CHN','ITA', 'IRN']]
```

```python
df5.columns = ['Brazil','United States', 'France', 'China', 'Italy', 'Iran']
```
