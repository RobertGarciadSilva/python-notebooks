### First Look to dataset Telecom

```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from sklearn.model_selection import train_test_split
```

### Overview

- Primeira coisa importante a perceber, geralmente é algo que não ocorre em dataset reais, é que o dataset aparenta não apresentar nenhum valor NA, ou seja, um caso bem ideal.

- A primeira coisa a ser feita será quebrar o dataset em dados para análise e treino (train) e dados para validação (test). Os dados serão quebrados de forma que 80% serão usados para treino e 20% para validação.

- A análise será feita a partir de então com os dados de treino (train).

Análisando os dados percebe-se que alguns dados NA estão representados por ' ', sendo então colocado essa observação na leitura do dataset abaixo.

```python
# read data
telecom = pd.read_csv('./splited_data/train.csv', na_values=' ')
```

```python
telecom.head()
```

```python
telecom.info()
```

Observa-se que a coluna <b>TotalCharges</b>, aparentemente, é a única coluna que apresenta valores faltantes, logo abaixo, é feito um rápido olhar para esses registros faltantes, tem-se que:
<lu>
    <li>A coluna <b>tenure</b>, que representa quantos meses a pessoa tem sido cliente da empresa, é nula.</li>
    <li>A coluna <b>Churn</b>, que reprsenta se a pessoa manteve ou não o contrado com a empresa apresenta todos valores 'No'.</li>
</lu>
Pelo apresentado, para tais clientes, em que a coluna <b>TotalCharges</b> apresenta NA, parece razoável consideramos que estão no primeiro mês de contrado com a empresa, não finalizando ainda esse mês, podendo então:
- Para os registros com valores NA na coluna "TotalCharges" será considerado como 0.

```python
telecom[telecom['TotalCharges'].isnull()]
```

```python
# Como discutido acima, valores NA em "TotalCharges" sendo passados para o valor númerico "0".
telecom['TotalCharges'].loc[lambda telcom: telecom['TotalCharges'].isnull()] = 0
```

```python
telecom.info()
```

As colunas <b>Unnamed: 0</b> e <b>customerID</b> não fornecem informações úteis para a análise, sendo então deletadas.

```python
telecom.drop(['Unnamed: 0','customerID'], axis=1, inplace=True)
```

```python
telecom = telecom.astype({'SeniorCitizen':'object'});
```

```python
telecom.info()
```

### Análise univariada

<b>gender</b>: Variável quantitativa nominal.

```python
telecom['gender'].value_counts(normalize=True)
```

Observa-se que não existe uma diferença grande entre as categorias "Male" e "Female" da presente coluna.

```python
telecom['SeniorCitizen'].value_counts(normalize=True)
```

<b>SeniorCitizen</b>: Variável qualitativa nominal. A maioria dos clientes é de não aposentados. 

```python
telecom['Partner'].value_counts(normalize=True)
```

<b>Partner</b>: Variável quantitativa nominal. Observa-se que não existe uma diferença muito grande entre a quantidade de clientes casados e não casados.

```python
telecom['Dependents'].value_counts(normalize=True)
```

<b>Dependents</b>: Variável quantitativa nominal. Observa-se que a maioria dos clientes possui dependentes.

```python

fig, (axe1, axe2) = plt.subplots(2, 1, figsize=(15,12))
fig.subplots_adjust(hspace=0.3)

telecom['tenure'].value_counts(normalize=True).sort_index().plot.bar(ax=axe1)
axe1.set_title('Barplot: Quantidade de meses contratados')
axe1.set_xlabel('Mês')
axe1.set_ylabel('Frequência absoluta')

axe2.hist(telecom['tenure'], edgecolor='black', bins=15)
axe2.set_title('Histograma: Quantidade de meses contratados')
axe2.set_xlabel('Número de meses')
axe2.set_ylabel('Frequência absoluta')


plt.show()
```

<b>Tenure</b>: Variável quantitativa discreta. 
- Observa-se um número grande de clientes, em torno de 8%, que mantiveram ou ainda possuiem contrato até o fim do primeiro mês.

Pode ser interessante análisar a realção entre <b>tenure</b> e <b>Churn</b>, respondendo perguntas como:
- Qual a porcentagem de clientes que finalizam o contrato até o fim do primeiro mês?
- Se a resposta do item anterior for grande, pode ser interessante se perguntar "Existe alguma variável ou conjunto de variáveis que influenciam o fim de contrato até o fim desse primeiro mês?".

```python

fig, axe = plt.subplots(3,3, figsize=(16,14))
fig.subplots_adjust(wspace=0.3)

telecom['PhoneService'].value_counts().plot.pie(ax=axe[0][0], autopct='%1.0f%%')
telecom['InternetService'].value_counts().plot.pie(ax=axe[0][1], autopct='%1.0f%%')
telecom['MultipleLines'].value_counts().plot.pie(ax=axe[0][2], autopct='%1.0f%%')
telecom['StreamingTV'].value_counts().plot.pie(ax=axe[1][1], autopct='%1.0f%%')
telecom['OnlineSecurity'].value_counts().plot.pie(ax=axe[2][2], autopct='%1.0f%%')
telecom['DeviceProtection'].value_counts().plot.pie(ax=axe[1][0], autopct='%1.0f%%')
telecom['OnlineBackup'].value_counts().plot.pie(ax=axe[1][2], autopct='%1.0f%%')
telecom['TechSupport'].value_counts().plot.pie(ax=axe[2][0], autopct='%1.0f%%')
telecom['StreamingMovies'].value_counts().plot.pie(ax=axe[2][1], autopct='%1.0f%%')


plt.show()
```

As variáveis: <b>PhoneService</b>, <b>InternetService</b>, <b>MultipleLines</b>, <b>StreamingTV</b>, <b>OnlineSecurity</b>, <b>DeviceProtection</b>, <b>OnlineBackup</b>, <b>TechSupport</b> e <b>StreamingMovies</b> são todos qualitativas nominais e representam os servições contratados pelos clientes, acima é mostrado uma rápida visualização da frequência relativa dos valores de cada uma dessas variáveis.

```python
telecom['Contract'].value_counts(normalize=True)
```

<b>Contract</b>: Variável qualitativa ordinal. Pode-se observar que mais da metade dos clientes, aproximadamente 54%, possuem o contrato "Month-to-month", ou seja, realizam o pagamento mês a mês.

```python
telecom['PaperlessBilling'].value_counts(normalize=True)
```

<b>PaperlessBilling</b>: Variável qualitativa nominal. Pode-se perceber que quase 60% dos clientes realizam pagamento sem fatura de papel. 

```python

telecom['PaymentMethod'].value_counts(normalize=True).plot.pie(autopct='%1.0f%%', title='PaymentMethod', figsize=(8,8), label='')
plt.show()
```

<b>PaymentMethod</b>: Variável qualitativa nominal, acima é mostrado a distribuição de frequência relativa dessa variável. 

```python
telecom['MonthlyCharges'].plot.hist(figsize=(10,6), title='Histograma: Monthly Charges', bins=15, density=True)
plt.xlabel('MonthlyCharges')
plt.xticks(np.arange(20,125,5))
plt.show()
```

<b>MonthlyCharges</b>: Variável quantitativa contínua. Observando o histograma acima percebe-se que a maior frequência relativa ocorre entre cobranças, aproximadas, de 20 a 25  (coniderando que esse valor é calculado em dollar).

```python
telecom['TotalCharges'].plot.hist(figsize=(12,6), bins=15, density=True, title='Histograma: Total Charges')
plt.xlabel('TotalCharges')
plt.xticks(np.arange(0,8500,500))
plt.show()
```

<b>total Charges</b>: Variável qualitativa contínua. Observa-se no histograma acima que a maior frequência relativa ocorre até 500$, sendo que aproximadamente 0.05% dos clientes tiveram um gasto total de até 500%.

```python
telecom['Churn'].value_counts(normalize=True).plot.pie(figsize=(8,8),autopct='%1.0f%%', title='Churn', label='')
plt.show()
```

<b>Churn</b>: Variável qualitativa nominal, <b>Label</b>. Acima obsreva-se que aproximadamente 73% dos clientes manteêm ou mantiveram o contrato com a Telecom.

```python
#Amostra de registrom que apresentam gastos mensais com a Telecom maior que a média de gastos.
sample_MonthlyChargesGreatherMean = telecom[telecom['MonthlyCharges'] > telecom['MonthlyCharges'].mean()]
```

```python
sample_MonthlyChargesGreatherMean.loc[:,['InternetService', 'TechSupport']].value_counts().plot.pie(autopct='%1.0f%%', label='', title='Pie: (InternetService, Techsupport)', figsize=(6,6))
plt.show()
```

Observa-se que:
- Os clientes que apresentam gastos maior que a média, <b>MonthlyCharges</b> maior que a média, apresentam algum serviço de internet, <b>InternetService</b>.<br>
- Os clientes que apresentam, em média, maior gasto mensal com a Telecom, MonthlyCharges, tendem a apresentam um maior cancelamento do contrado em comparação aos que apresentam, em média, menor gasto.

```python
sns.boxplot(y=sample_MonthlyChargesGreatherMean['tenure'], x=sample_MonthlyChargesGreatherMean['Churn'])
```

```python
# Exemplo: Aplicando TEOREMA LIMITE CENTRAL

x_med_am = []

for i in range(500):
    x = telecom['MonthlyCharges'].sample(60).mean()
    x_med_am.append(x)
    
med = pd.DataFrame({'Médias':x_med_am})
med.hist(density=True)
```

```python

print(med.std())
print(med.mean())
```

```python
print(telecom['MonthlyCharges'].std()/np.sqrt(60))
print(telecom['MonthlyCharges'].mean())
```

```python

from scipy.stats import norm

norm.pdf(65, loc=telecom['MonthlyCharges'].mean(), scale=telecom['MonthlyCharges'].std()/np.sqrt(60))
norm.cdf(65, loc=telecom['MonthlyCharges'].mean(), scale=telecom['MonthlyCharges'].std()/np.sqrt(60))
```

```python
e = norm.interval(0.6, loc=telecom['MonthlyCharges'].mean(), scale=telecom['MonthlyCharges'].std()/np.sqrt(60))

med.hist(density=True)
plt.vlines(e[0], plt.gca().get_ylim()[0], ymax=plt.gca().get_ylim()[1], color='red', lw=3)
plt.vlines(e[1], ymin=plt.gca().get_ylim()[0], ymax=plt.gca().get_ylim()[1], color='red', lw=3)
```
