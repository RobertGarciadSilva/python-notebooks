## SimpleExampleDecisionTreeClassifier

```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_blobs
```

```python

X, y = make_blobs(n_samples=450, centers=3, cluster_std=3, random_state=42)
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python

def visualize_treeClassifier2D(model_tree, X, y, axe=None, cmap='viridis'):
    '''
    usado para visualização, em um sistema de eixos 2D, das fronteiras 
    DecisionTreeClassifier
    '''
    
    axe = axe or plt.gca()
    axe.scatter(X[:,0], X[:,1], c=y, zorder=3, cmap=cmap, edgecolor='black') # zorder: Dentre os valores possíveis {1,2,3}, pode ser usado para controlar a ordem da plotagem
    #axe.axis('tight')
    axe.axis('off')
    xlim = axe.get_xlim()
    ylim = axe.get_ylim()
    
    model_tree.fit(X, y)
    
    # np.linspace(xlim[0],xlim[1]) it is equivalente to np.linspace(*xlim)
    x_grid, y_grid = np.meshgrid(np.linspace(xlim[0], xlim[1], 200), np.linspace(ylim[0],ylim[1], 200))
    
    #criando pares ordenados (x,y) entre x_grid e y_grid
    parOrd_xy = np.vstack([x_grid.ravel(), y_grid.ravel()]).T 
    
    Z = model_tree.predict(parOrd_xy).reshape(x_grid.shape)
   
    axe.contourf(x_grid, y_grid, Z, alpha=0.5, zorder=1)
    
    
    plt.show()


    
    
```

```python

model_tree = DecisionTreeClassifier(max_depth=4)


fig = plt.figure(figsize=(10,6))    
axe = plt.gca()

visualize_treeClassifier2D(model_tree, X, y, axe=axe)
```
