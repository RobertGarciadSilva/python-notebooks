```python

#Import libraries

import os
import zipfile
from six.moves import urllib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
```

```python

# Caso não tenha os dados baixados a presente célula pode ser usada para criar o diretorio, baixar e salvar os dados.

DEFAULT_PATH_URL_ZIP = 'https://github.com/RobertGarciadSilva/python-notebooksFiles/blob/master/notebooks/dataCaliforniaHousingPrices/data/california-housing-prices.zip?raw=true'
DEFAULT_PATH_URL_CSV = 'https://raw.githubusercontent.com/RobertGarciadSilva/python-notebooksFiles/master/notebooks/dataCaliforniaHousingPrices/data/housing.csv'

DEFAULT_DIR_SAVE_DATA = 'data_CaliforniaHousingPrices'



def get_data_californiaHousingPricesZip(path_url = DEFAULT_PATH_URL_ZIP, dir_save_data = DEFAULT_DIR_SAVE_DATA):
    '''
    Obtem arquivo Zip, com os dados, a partir da url fornecida acima, extraindo os arquivos presentes no zip. Os dados são salvos em um diretorio, criado no diretorio atual, chamado "data_CaliforniaHousingPrices".
    Exemplo read: pd.read_csv('./data_CaliforniaHousingPrices/housing.csv')
    '''
    if(os.path.isdir(dir_save_data) == False):
        os.makedirs(dir_save_data)
    
    zip_name = 'california-housing-prices.zip'
    zip_path = os.path.join(dir_save_data, zip_name)
    urllib.request.urlretrieve(path_url, zip_path)
    getZipFile = zipfile.ZipFile(zip_path)
    getZipFile.extractall('./'+dir_save_data)
    getZipFile.close()
    

    
def get_data_californiaHousingPricesCSV(path_url = DEFAULT_PATH_URL_CSV, dir_save_data = DEFAULT_DIR_SAVE_DATA):
    '''
    Obtem arquivo CSV, com os dados, a partir da url fornecida acima, os dados são salvos em um diretorio, criado no diretorio atual, chamado "data_CaliforniaHousingPrices".
    Exemplo read: pd.read_csv('./data_CaliforniaHousingPrices/housing.csv')
    '''
    if(os.path.isdir(dir_save_data) == False):
        os.makedirs(dir_save_data)
        
    with urllib.request.urlopen(path_url) as response:
        data = response.read()
        
    with open('./'+ dir_save_data + '/california-housing-prices.csv','wb') as file:
        file.write(data)
    
    # Os dois "With" acima podem ser colocados na mesma linha, ficando:
    
    # with urllib.request.urlopen(path_url) as response, open('./'+ dir_save_data + '/california-housing-prices.csv', 'wb') as out_file:
    #    data = response.read() # a `bytes` object
    #    out_file.write(data)
 


#get_data_californiaHousingPricesZip()    
#get_data_californiaHousingPricesCSV()
```

```python
valCountsOceanProximity = data['ocean_proximity'].value_counts(normalize=True); valCountsOceanProximity
```

```python

# substitui Nan na coluna "total_bedrooms" por sua mediana

total_bedroomsMedian = data['total_bedrooms'].median()

def total_bedroomsToMean(x):
    if(pd.isnull(x) == True):
        x = total_bedroomsMedian
    return x

data['total_bedrooms'] = data['total_bedrooms'].apply(total_bedroomsToMean)
```

```python

# categorizando "ocean_proximity"

oceanProximityKeys = valCountsOceanProximity.keys()

def cat_oceanProximity(x):
    for i, val in enumerate(oceanProximityKeys):
        if(x == val):
            x = i
            break
    return x

data['ocean_proximity'] = data['ocean_proximity'].apply(cat_oceanProximity)
    
```

```python
data['ocean_proximity'].value_counts()
```

```python

label = 'median_house_value'

x = np.array(data.drop([label,'ocean_proximity'], axis='columns'))

y = np.array(data[label])
```

```python

n = 1000 # número de tentativas
bestModel = 0
bestAccuracy = 0

for i in range(n):
    x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.2)

    model = linear_model.LinearRegression()
    model.fit(x_train, y_train)
    accuracy = model.score(x_test, y_test)
    
    if(accuracy > bestAccuracy):
        bestAccuracy = accuracy
        bestModel = model

print('Accuracy: ', bestAccuracy)
```

```python

# gera arquivo para ir guardando os valores da acuracia obtida ao longo da evolução do trabalho

import time


def save_currentAccuracy():
    if(os.path.isdir('historyAccuracy') == False): #Verifica se existe um diretorio para salvar a acurancia, não existindo é criado.
        os.makedirs('historyAccuracy')
    
    cabecalho = 80*'-' + '\nDate' + 30*' ' + 'Accuracy\n' + 80*'-' + '\n' # string contendo cabeçalho para o arquivo
    timeSeconds = time.time() #sengundos desde epoch
    localTime = time.ctime(timeSeconds) # Obtem data e hora atual baseado nos segundos desde epoch
    
    # verifica se existe um arquivo contendo as informações, não havendo é criado esse arquivo com o cabeçao definido acima.
    if(os.path.isfile('./historyAccuracy/history-accuracy.txt') == False):
        with open('./historyAccuracy/history-accuracy.txt', 'w+') as file:
            file.write(cabecalho)
    
    
    #Salva no arquivo a data atual, obtida logo acima, e a última acuracia salva em "bestAccuracy".
    with open('./historyAccuracy/history-accuracy.txt', 'a+') as file:
        file.write(localTime); file.write(10*' ')
        file.write(str(bestAccuracy)); file.write('\n')

        
#save_currentAccuracy()
```
