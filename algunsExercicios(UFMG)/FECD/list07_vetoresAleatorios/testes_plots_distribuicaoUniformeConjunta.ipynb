## Examples

```python

import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

from scipy.stats import uniform
```

```python

num = 2000

X = uniform.rvs(size=num)
Y = uniform.rvs(size=num)


plt.hist2d(X, Y, density=True)
plt.show()

```

```python

fig = plt.figure(figsize=(9,7))
axe = fig.add_subplot(projection='3d')
hist, x_bins, y_bins = np.histogram2d(X, Y, range=[[0,1],[0,1]], bins=6, density=True)

xGrid, yGrid = np.meshgrid(x_bins[:-1] + 0.5, y_bins[:-1] + 0.5)
xGrid = xGrid.ravel()
yGrid = yGrid.ravel()
zpos = 0

dx = dy = 0.5 * np.ones_like(zpos)
dz = hist.ravel()

axe.bar3d(xGrid, yGrid, zpos, dx, dy, dz, zsort='average', color='red', edgecolor='black')

plt.show()
```
