```python

import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

<b>K</b> pode ser obtido fazendo:<br>

```python

def densidadeConjunta_fxy(x,y):
    k = 2/3
    if((x < 0 and x > 1) and (y < 0 and y > 1)):
        return 0
    else:
        fxy = k*(x + y + x*y)
        return fxy
```

```python

x = np.arange(0,1,0.1)
y = np.arange(0,1,0.1)
xGrid, yGrid = np.meshgrid(x,y)

Z = (2/3)*(xGrid + yGrid + xGrid*yGrid)
#Z = np.array(densidadeConjunta_fxy(np.ravel(xGrid), np.ravel(yGrid))).reshape(xGrid.shape)

fig = plt.figure(figsize=(12,8))
axe = fig.gca(projection='3d')

surf = axe.plot_surface(xGrid, yGrid, Z, cmap='viridis', edgecolor='none')
axe.set_xlabel('X')
axe.set_ylabel('Y')
axe.set_zlabel('Z')
axe.set_title('Plot surface fxy(X,Y)', size=15)

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

A densidade marginal <b>fx(X)</b> pode ser obtida fazendo:
<img src='imgs/q174.jpeg' style='width:400px'>

```python

def densidadeMarginal_fx(x):
    k = 2/3
    return (k/2)*(3*x + 1)
```

```python


x = np.arange(0,1,0.1)
y = densidadeMarginal_fx(x)

x_avaliar = np.array([0.2,0.5]) # avaliando densidade marginal fx(X) nesses pontos
y_avaliar = densidadeMarginal_fx(x_avaliar)

fig, axe = plt.subplots()
axe.plot(x,y)
axe.scatter(x_avaliar[0], y_avaliar[0], color='red', label='fx(X = {0}) = {1:.3f}'.format(x_avaliar[0],y_avaliar[0]))
axe.scatter(x_avaliar[1], y_avaliar[1], color='green', label='fx(X = {0}) = {1:.3f}'.format(x_avaliar[1],y_avaliar[1]))
axe.set_xlabel('X')
axe.set_ylabel('fx(X)')
axe.set_title('Densidade marginal fx(X)')
axe.grid(alpha=0.7)
axe.legend()

plt.show()
```

```python

def fxCondicionadoY(x, y):
    return 2(x + y + x*y)/(3*y + 1)
```
