```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns; sns.set()

from scipy.stats import norm
from scipy.stats import uniform

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

```

### Gerando amostra de dados 01

```python

# Gerando atributos

x1 = np.arange(1,25) # primeiro atributo
x2 = np.round(25*uniform.rvs(size=24),1) # segundo atributo

print('Atributo x1:\n', x1, end='\n\n')
print('Atributo x2:\n', x2)
```

```python

# Gerando matrix com os dados que serão usados para gerar Y

X = np.array([np.ones(len(x1)), x1, x2])
X = np.vstack(X)
X = X.T; X
```

```python

# vetor de coeficientes
beta = [1,1.5,0.7]
```

```python

mu = np.matmul(X, beta) # Y = 1 + 1.5x1 + 0.7x2
sigma = 5 # desvio padrão dos dados de erro
epsilon = norm.rvs(size=len(x1), loc=0, scale=sigma) # vetor com o erro associado, definido por N(0,σ2= 25)

y = mu + epsilon # dados gerados
```

```python
# dataframe com os dados
data = pd.DataFrame(np.array([y, x1, x2]).T, columns=['y','x1','x2'])
data.head(5)
```

```python

# plotagem dos dados gerados
sns.pairplot(data)
plt.show()

```

```python
x_train = np.array([x1,x2]).T; #atribudo de entrada para o modelo
```

```python

# aplicando modelo de regressão linear
model_linear = LinearRegression().fit(x_train, y)
```

```python

#coeficientes obtido pelo modelo
print('Coef. angular obtido pelo modelo: ', model_linear.coef_)
print('Coef. linear obtido pelo modelo: ', model_linear.intercept_)
```

```python

# obtendo y predito pelo modelo de regressão linear
coef_model_obtido = np.array([model_linear.intercept_, model_linear.coef_[0], model_linear.coef_[1]])
y_model = np.matmul(X, coef_model_obtido)
```

```python

plt.title('y vs y predito')
plt.scatter(y_model, y)
plt.xlabel('y predito')
plt.ylabel('y')
plt.show()
```

```python
# matriz de correlação entre y predicted e y_true
matCorr = np.corrcoef([y_model,y]); matCorr
```

```python
matCorr**2
```

```python
# r2_score
rSquared = r2_score(y, y_model); rSquared
```

Observe que o quadrado da correlação obtida entre y_true e y_predicted equivale ao valor obtido por r2_score.

### Gerando n amostras de dados

```python

nAmostras = 1000
size_amostras = 24;

coefBeta = [] # armazena os coeficientes obtidas por cada modelo

for a in range(nAmostras):
    norm_rvs = norm.rvs(size=size_amostras, loc=0, scale=sigma)
    y = mu + norm_rvs
    
    model_linear = LinearRegression().fit(x_train, y)
    
    coefBeta.append([model_linear.intercept_, model_linear.coef_[0], model_linear.coef_[1]])
    
```

```python
# transformando coefBeta em um np.array e remodelando para um novo formato de dimensão
coefBeta = np.array(coefBeta).reshape(-1,3); #coefBeta[:,1]
```

```python

# plotagem de resultados

fig, ([axe1,axe2],[axe3,axe4]) = plt.subplots(2,2, figsize=(14,10))
fig.subplots_adjust(hspace=0.4)

axe1.hist(coefBeta[:,0], density=True, edgecolor='black') # distribuição dos coeficientes lineares
axe1.vlines(beta[0], 0, axe1.get_ylim()[1], lw=3, color='red', label='Coef. usado para\ngerar as amostras') # real valor do coef. angular, usado para gerar as amostras
axe1.set_title('Distribuição dos coeficientes lineares gerados pelo modelo')
axe1.set_xlabel('Beta[0], coef. Linear')
axe1.set_ylabel('Frequency')
axe1.legend()

axe2.hist(coefBeta[:,1], density=True, edgecolor='black') # distribuição do coeficiente angular x1
axe2.vlines(beta[1], 0, axe2.get_ylim()[1], lw=3, color='red', label='Coef. usado para\ngerar as amostras') # real valor do coef. linear, usado para gerar as amostras
axe2.set_title('Distribuição do coeficiente Beta[1] gerado pelo modelo')
axe2.set_xlabel('Beta[1], coef. Angular')
axe2.set_ylabel('Frequency')
axe2.legend()

axe3.hist(coefBeta[:,2], density=True, edgecolor='black') # distribuição do coeficiente angular x2
axe3.vlines(beta[2], 0, axe2.get_ylim()[1], lw=3, color='red', label='Coef. usado para\ngerar as amostras') # real valor do coef. linear, usado para gerar as amostras
axe3.set_title('Distribuição do coeficiente Beta[2] gerado pelo modelo')
axe3.set_xlabel('Beta[1], coef. Angular')
axe3.set_ylabel('Frequency')
axe3.legend()

axe4.scatter(coefBeta[:,1],coefBeta[:,2], edgecolor='black') # scatter plot
axe4.set_title('Scatter plot: Coef. angular Beta[2] vs Coef. angular Beta[1], amostras geradas')
axe4.set_xlabel('Beta[1], coef. Angular Beta[1]')
axe4.set_ylabel('Beta[2], coef. Angular Beta[2]')


plt.show()

```
