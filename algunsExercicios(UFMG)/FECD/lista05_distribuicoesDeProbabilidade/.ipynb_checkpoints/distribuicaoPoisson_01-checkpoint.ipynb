## Simple example using Poisson

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import poisson
```

```python

# some useful function
# necessário o uso de: from import scipy.stats import poisson

def poisson_MaxPmf(param, values):
    """Find the max value P(X = x), pmf (probabilistic mass function), of Poisson distribuition,
    using param. Return the value x and P(X = x) 
    param: Poisson param
    values: Values x to calculate P(X = x)
    return: Value x and P(x = x) to max P(X = x)
    
    Example:
    x = range(0,21+1) #intervalo arbitrário
    x_max, prob_max = poisson_MaxPmf(mi_1, x)
    
    Output:
    x_max = 0
    prob_max = 0.48190899009020244
    
    """
    maxi, ind = 0, 0
    for b in values:
        p = poisson(param).pmf(b)
        if(p > maxi):
            maxi = p
            ind = b
    return (ind, maxi)

def prob_interval_poisson(param, a,b):
    """Obtém a probabilidade dentro de um intervalo [a,b] para a distribuição de poisson
    param: int, float, parãmetro da distribuição de poisson
    [a,b]: int, intervalo para o cálculo da probabilidade total
    return: probabilidade total no intervalo [a,b]
    
    
    """
    sum_prob = 0
    for i in range(a,b+1):
        sum_prob += poisson(param).pmf(i)
    return sum_prob
```

```python

# instancing Poisson
x = range(0,21+1) #intervalo arbitrário
mi_1 = 0.73
mi_2 = 10
poisson_pmf1 = poisson(mi_1).pmf(x) # probabilistic mass function
poisson_cdf1 = poisson(mi_1).cdf(x) # cumulative distribution function

poisson_pmf2 = poisson(mi_2).pmf(x) # probabilistic mass function
poisson_cdf2 = poisson(mi_2).cdf(x) # cumulative distribution function

# plots
fig, ([axe1,axe2],[axe3,axe4]) = plt.subplots(2,2, figsize=(15,9))
fig.subplots_adjust(hspace=0.4)

axe1.vlines(x, 0, poisson_pmf1, lw=3, label='P(X = x)')
axe1.set_title('Poisson - Probabilistc mass function, param = {0}'.format(mi_1))
#axe1.set_yticks(poisson_pmf1)
axe1.set_xticks(x)
axe1.grid(alpha=0.6)
axe1.set_xlabel('x')
axe1.set_ylabel('Probabilistic mass function')

x_max, prob_max = poisson_MaxPmf(mi_1, x)
axe1.vlines(x_max, 0, prob_max, lw=2, color='red', label='Max(P(X = x))')

axe1.legend()

axe2.step(x, poisson_cdf1, lw=3)
axe2.set_title('Poisson - Cumulative distribution function, param = {0}'.format(mi_1))
axe2.set_xticks(x)
axe2.grid(alpha=0.6)
axe2.set_xlabel('x')
axe2.set_ylabel('Cumulative function')

axe3.vlines(x, 0, poisson_pmf2, lw=3, label='P(X = x)')
axe3.set_title('Poisson - Probabilistc mass function, param = {0}'.format(mi_2))
#axe2.set_yticks()
axe3.set_xticks(x)
axe3.grid(alpha=0.6)
axe3.set_xlabel('x')
axe3.set_ylabel('Probabilistic mass function')

x_max2, prob_max2 = poisson_MaxPmf(mi_2, x)
axe3.vlines(x_max2, 0, prob_max2, lw=2, color='red', label='Max(P(X = x))')

axe3.legend()

axe4.step(x, poisson_cdf2, lw=3)
axe4.set_title('Poisson - Cumulative distribution function, param = {0}'.format(mi_2))
axe4.set_xticks(x)
axe4.grid(alpha=0.6)
axe4.set_xlabel('x')
axe4.set_ylabel('Cumulative function')

plt.show()


```

- Usando ppois do R, calcule P(a ≤ X ≤ b)

Usando a função definida acima <b>prob_interval_poisson</b>. Temos:

```python
p1 = prob_interval_poisson(0.73, 0,1)
p2 = prob_interval_poisson(10, 6,13)
print('P(0 <= X <= 1), param = 0.73: ', p1)
print('P(6 <= X <= 13), param = 10: ', p2)
```
