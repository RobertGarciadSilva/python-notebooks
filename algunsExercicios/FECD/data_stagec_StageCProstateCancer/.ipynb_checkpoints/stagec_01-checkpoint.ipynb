## A simpleExample

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; #sns.set()

from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python

data = pd.read_csv('./data/stagec.csv')
```

```python

data.head()
```

```python
data.info()
```

```python
data.describe()
```

## Quicly clear data

```python

data.dropna(inplace=True)
#data.index = np.arange(0,len(data)) # reindex
```

```python
data['ploidy'].value_counts()
```

```python

def transformPloidy(x):
    if(x == 'diploid'):
        x = 1
    elif(x == 'tetraploid'):
        x = 2
    else:
        x = 3
    return x
        
data['ploidy'] = data['ploidy'].apply(transformPloidy)
```

```python
data.head()
```

```python


def transform_pgstat(x):
    if(x == 0):
        x = 'No'
    elif(x == 1):
        x = 'Prog'
    return x

data['pgstat'] = data['pgstat'].apply(transform_pgstat)
```

## Using Decision Tree Classifier

```python

X = data.drop(['pgstat','pgtime','Unnamed: 0'], 1)
y = data['pgstat']
```

```python
X.head()
```

```python
y.head()
```

```python
y.value_counts()
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=35, train_size=0.8)
```

```python
y_train.value_counts()
```

```python

# outra forma de separar os dados em treino e teste
#rns = np.random.RandomState(35)
#indx = rns.randint(min(data.index),max(data.index), int(0.8*len(data)))

#data_train = data.iloc[indx]; #data_train

#X_train, y_train = data.drop(['pgstat','pgtime','Unnamed: 0'],1), data['pgstat']
```

```python

model_decisionTree = DecisionTreeClassifier(max_depth=5, max_leaf_nodes=5)
model_decisionTree.fit(X_train, y_train)


#passando outros parametros
#model_decisionTree = DecisionTreeClassifier(max_depth=3, min_samples_leaf=6, max_leaf_nodes=5)
```

```python
model_decisionTree.classes_
```

```python
model_decisionTree.feature_importances_
```

```python

fig = plt.figure(figsize=(18,12))

plot_tree(model_decisionTree, filled=True, feature_names=X_train.columns, class_names=['No','Prog'])
plt.title('Plot tree', size=20)
plt.show()
```

```python

# confusion matrix

y_pred = []
X = np.array(X_test)

for i in range(len(X)):
    y_pred.append(model_decisionTree.predict(X[i].reshape(-1,6)))

confusion_mat = confusion_matrix(y_test.values, y_pred); confusion_mat
```

```python

# confusion matrix in heatmap

sns.heatmap(confusion_mat, square=True, annot=True, cbar=False, fmt='d', xticklabels=['No','Prog'], yticklabels=['No','Prog'])
plt.xlabel('True label')
plt.ylabel('Predicted label')
plt.title('Confusion Matrix')
plt.show()
```

```python
# classification_report
print(classification_report(y_test.values, y_pred, target_names=['No','Prog']))
```

```python
# refazendo a rvore usando outra amostra dos dados, usando outra seed para dividir as amostras de treino e test

# dividindo a amostra em treino e test
X = data.drop(['pgstat','pgtime','Unnamed: 0'], 1)
y = data['pgstat']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, train_size=0.8)

# gerando e aplicando o modelo sobre os dados
model_decisionTree = DecisionTreeClassifier(max_depth=5, max_leaf_nodes=5)
model_decisionTree.fit(X_train, y_train)


# plotando as rvores
fig = plt.figure(figsize=(18,12))

plot_tree(model_decisionTree, filled=True, feature_names=X_train.columns, class_names=['No','Prog'])
plt.title('Plot tree 2', size=20)
plt.show()
```

## Questions

- Quais  variaveis  foram  usadas  e  quais  nao  foram  usadas  na   arvore  obtida  pela  segmentao recursiva do <b>rpart</b>?

No exemplo feito acima, como pode-se ver na imagem gerada em <b>Plot tree</b> as variveis usadas foram: g2 e gleason.
As variveis no usadas portanto: age, eet, grade e ploidy. 
A diferena entre a rvore gerada no presente exemplo e a apresentada no livro pode dever-se a instabilidade do mtodo, diferena das amostras para gerar a rvore, podendo estar ligadas ao tratamento dos dados antes da aplicao do mtodo em si, por exemplo, remoo de determinadas linhas que apresentam valores NA, como feito acima.

<br><br>
- Forneca estimativas de probabilidades condicionais para as folhas da  arvore.  Isto  e, forneca uma  estimativa  numerica  para P(Y=  1|X1,X2,...)  em  cada  folha  (no  terminal)  da   arvore especificando o conjunto de atributosXs e seus valores em cada no final

Observando a plotagem da rvore feito acima, em <b>Plot tree</b>, podemos tirar algumas rpidas estimativas de probabilidades condicionais:<br>
P(Y = 1 (Prog) | g2 <= 3.81) = 1<br>
P(Y = 1 (Prog) | g2 <= 22.675, gleason <= 5.5) = 22/25

<br><br>
- V ou F: 
Como a eet nao apareceu na  arvore, isto sugere (pois nao  e uma prova definitiva) que:
(a) P(Y= 1) = P(Y= 1|eet = yes) = P(Y= 1|eet = no);
(b) Os eventos eet = yes e eet = no sao independentes do evento Y = 1.

Para esse caso em particular, levando em conta a amostra usada, e observando o sudo acima do mtodo <b>model_decisionTree.feature_importances_</b>, que apresenta a importncia dos atributos na rvore, pode-se sugerir-se ento que o atributo <b>eet</b>  independente do evento Y = 1, fazendo com que P(Y= 1|eet = yes) = P(Y= 1|eet = no) = P(Y = 1), tornado-se ambas as alternativas VERDADEIRAS, mas, novamente, observe que essa sugesto aplica-se apenas a esse caso particular, como as rvores de deciso so bastante instveis para pequenas amostras, no se pode concluir isso em definitivo.
