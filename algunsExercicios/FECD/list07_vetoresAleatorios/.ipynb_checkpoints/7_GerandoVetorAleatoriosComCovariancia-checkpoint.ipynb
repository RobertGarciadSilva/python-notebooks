## Examples

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import norm
```

<img src='imgs/q71.jpg' style=''>

<img src='imgs/q72.jpg' style=''>

```python

num = 2000 # numero de pontos em cada v.a. do vetor aleatorio
dim = 3 # dimensões do vetor aleatório
matCov = np.array([[4,9,-14],[9,30,-44],[-14,-44,94]]) #matrix de covariancia
mu = np.array([10,20,-50]).reshape(3,1) # média dos dados de cada dimensão do vetor aleatório

amostra = np.array([]) # vetor aleatório

for d in range(dim):
    Yd = norm.rvs(size=num) # para cada dimensão do vetor aleatório gero num observações com distribuição normal, N(0,1)
    amostra = np.append(amostra, Yd)

amostra = amostra.reshape(3,-1) #vetor aleatório de dim dimensões, Y = (X1, X2, X3)
# Repare que nesse ponto cada Xi do vetor aleatório é independente e identicamente distribuido,
# cada Xi, dimensão, segue também a distribuição normal (0,1)


# Plotagem de cada dimensão do vetor aleatório
fig, (axe1, axe2, axe3) = plt.subplots(1,3, figsize=(16,4))
fig.subplots_adjust(wspace=0.3)

axe1.hist(amostra[0], density=True, edgecolor='black', color='red')
axe1.set_xlabel('x')
axe1.set_ylabel('Relative frequency')
axe1.set_title('Histograma de X{0} do vetor aleatório Y = (X1,X2,X3)'.format(1))

axe2.hist(amostra[1], density=True, edgecolor='black', color='green')
axe2.set_xlabel('x')
axe2.set_ylabel('Relative frequency')
axe2.set_title('Histograma de X{0} do vetor aleatório Y = (X1,X2,X3)'.format(2))

axe3.hist(amostra[2], density=True, edgecolor='black')
axe3.set_xlabel('x')
axe3.set_ylabel('Relative frequency')
axe3.set_title('Histograma de X{0} do vetor aleatório Y = (X1,X2,X3)'.format(3))

plt.show()
```

```python

L = np.linalg.cholesky(matCov) # usando a decomposição de cholesky, da algebra linear, para obter L, 
# de modo que L*(L,transposta) = matCov, veja referência [2], consulte o livro em [4], capítulos:
# vetores aleatórios e distribuição normal multivariada

X = np.add(mu, np.matmul(L,amostra)) # somando as médias ao produto das matrizes L por amostra
```

Observe que agora X é um vetor aleatório de três dimensões <b>X = (X1, X2, X3)</b>, com cada dimensão tendo as respectivas médias aproximadas iguais a matríx mu = [10,20,-50], e possuindo uma correlação, aproximada, entre si dada pela matrix de covariancia em matCov. 
<br>Verificamos essas correspondências abaixo:

```python
print('Média da v.a. X{0} do vetor aleatório X = (X1, X2, X3): {1:.3f}'.format(1, np.mean(X[0])))
print('Média da v.a. X{0} do vetor aleatório X = (X1, X2, X3): {1:.3f}'.format(2, np.mean(X[1])))
print('Média da v.a. X{0} do vetor aleatório X = (X1, X2, X3): {1:.3f}'.format(3, np.mean(X[2])))
# compare com mu = [10,20,-50]
```

A diagonal da matrix de covariancia corresponde a variancia de cada dimensão, logo, podemos verifica-la usando:

```python
print('Variância da v.a. X{0} do vetor aleatório X = (X1, X2, X3): {1:.3f}'.format(1, np.std(X[0])**2))
print('Variância da v.a. X{0} do vetor aleatório X = (X1, X2, X3): {1:.3f}'.format(2, np.std(X[1])**2))
print('Variância da v.a. X{0} do vetor aleatório X = (X1, X2, X3): {1:.3f}'.format(3, np.std(X[2])**2))
# compare com a diagonal da matríx de covariancia, matCov = [4,30,94]
```

A matrix do covariancia para os dados em X pode ser calculada rapidamente usando:

```python
np.cov(X)
#Compare com a matrix de covariancia, matCov, que usamos para gerar os dados em X.
```

Uma matrix de correlação pode ser plotada também, como segue abaixo:

```python

axe = sns.heatmap(np.corrcoef(X), annot=True, linewidth=0.2, xticklabels=['X1','X2','X3'], yticklabels=['X1','X2','X3'])
axe.set_title('Matrix de correlação entre X = (X1,X2,X3)')
plt.show()
```

Abaixo plotamos scatter plots de cada variavel de X em comparação com si mesma, em que é exibido histogramas, e com as outras variaveis. Rápidamente podemos comparar a relação entre a matrix de correlação e as plotagens abaixo. Lembrando da definição de correlação, <b>coeficiente de correlação de Pearson</b>, que pode ser verificado em [5] com mais detalhes mas que aqui usamos uma pequena parte, como: <br><br>
"O coeficiente de correlação de Pearson é +1 no caso de uma relação linear (correlação) direta perfeita (crescente), -1 no caso de uma relação linear (anticorrelação) decrescente perfeita (inversa) e qualquer valor no intervalo aberto (-1, 1) nos outros casos indicando o grau de dependência linear entre as variáveis."<br><br>
Ou seja, supondo uma v.a. X1 e X2, em que X1 cresce quando X2 também crese, o coeficiente de Pearson será positivo e será tanto mais próximo de 1 quanto mais linearmente dependentes são X1  e X2, 0 representa nenhuma relação linear e -1 representa uma relação linear indicando que quando X1 cresce, X2  decai.
Para mais detalhes, consulte [4], cap.(s): Vetores aleatórios e distribuição normal multivariada.

Abaixo estimamos a Correlação entre X1 e X2, manualmente, usando a definição em [5], como:

<img src='imgs/q73.jpg' style=''>
<img src='imgs/q74.jpg' style=''>

```python


def covariancia(X, Y):
    muX = np.mean(X)
    muY = np.mean(Y)
    cov = 0
    for x, y in zip(X, Y):
        cov += (x - muX)*(y - muY)
    return cov/len(X)

covX1X2 = covariancia(X[0], X[1])
corrX1X2 = covX1X2/(np.std(X[0])*np.std(X[1]))

print('Covariancia X1,X2 de X = (X2,X2,X3) = {0:.3f}'.format(covX1X2))
print('Correlação X1,X2 de X = (X2,X2,X3) = {0:.3f}'.format(corrX1X2))
#Compare acima
```

Uma outra forma de escrever o cálcula para a correlação, sendo mais intuitiva encontra-se abaixo, retirada de [4]:

<img src='imgs/q75.jpg' style=''>

ou seja, o coeficiente de correlação de Pearon entre as v.a. X1 e X2 pode ser calculado tomando a esperança do produto entre as v.a. X1 e X2, ambas padronizadas, cálculado o desvio padrão.

```python

def padroniza_padrao(X):
    mu = np.mean(X)
    std = np.std(X)
    X_padrao = []
    for x in X:
        X_padrao.append(x - mu)
    return X_padrao/std

corrX1X2 = np.mean(padroniza_padrao(X[0])*padroniza_padrao(X[1])); 

print('Correlação X1,X2 de X = (X2,X2,X3) = {0:.3f}'.format(corrX1X2))
```
