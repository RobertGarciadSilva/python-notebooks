```python

import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import uniform
from scipy.stats import binom
from scipy.stats import gompertz
```

Definição da função acumulada de uma função de desidade de probabilidade:
<img src='imgs/cumulativeDefinition.jpg' style=''>
Reference: [4]

```python

# definindo f(x) e a inversa da sua função acumulada 

def fx(x):
    return (3/2)*np.sqrt(x)


#cumulative f(x) => f(x) = x**(3/2)
# inversa cumulative f(x) => x = u**(2/3)

def inversa_cumulative_fx(u):
    return u**(2/3)
    
```

```python

# O objetivo é gerar dados para uma função densidade de probabilidade arbitrária, usando
# a distribuição uniforme [0,1], esse método é chamado de transformada inversa

#gerando 10000 valores com distribuição uniforme entre [0,1]
num = 10000
randUniformVa = uniform.rvs(size=num)

#usado para plotar a função densidade de probabilidade de f(x)
x = np.linspace(0,1,num)
y = fx(x)

#dados gerados para a distribuição alvo, f(x), usando a inversa da função acumulada de f(x) 
#com entrada valores da distribuição uniforme [0,1]
dados_distribuicao_fx = inversa_cumulative_fx(randUniformVa)


fig, axe = plt.subplots(figsize=(10,5))
axe.hist(dados_distribuicao_fx, density=True, edgecolor='black', label='Histograma dados gerados  para distribuição f(x)')
axe.plot(x, y, color='red', lw=2, label='Função densidade de probabilidade \n f(x) = (3/2)*np.sqrt(x)')
axe.legend()
axe.set_xlabel('x')
axe.set_ylabel('Frequência/densidade de probabilidade para f(x)')
axe.set_title('Dados gerados vs densidade de probabilidade para função f(x), , USANDO O MÉTODO DA TRANSFORMADA INVERSA')

plt.show()
```

```python

num = 10000
M = max(y) # constante M para multiplicar g(x)
U = uniform.rvs(size=num) # distribuição da qual sei amostrar, nesse caso foi escolhido a distribuição uniforme.


fig, (axe0, axe1) = plt.subplots(2,1, figsize=(12,10))
fig.subplots_adjust(hspace=0.3)


# os passos para o método aceitação rejeição estão nesse for
X = []
for Y in U:
    razao = fx(Y)/M # como estou utilizando um distribuição uniforme para a distribuição que sei amostrar, isso equivale fx(Y)/M*uniform.pdf(Y), Mg(x), observe o segundo plot
    aceita = binom(1,razao).rvs(size=1) # jogando uma moeda para cima com probabilidade "razao" para ver se Y será pego ou não para a distribuição alvo
    if(aceita == 1):
        X.append(Y)
    

axe0.hist(X, density=True, edgecolor='black', label='Histograma dados gerados  para distribuição f(x)')
axe0.plot(x, y, color='red', lw=2, label='Função densidade de probabilidade \n f(x) = (3/2)*np.sqrt(x)')
axe0.legend()
axe0.set_xlabel('x')
axe0.set_ylabel('Frequência/densidade de probabilidade para f(x)')
axe0.set_title('Dados gerados vs densidade de probabilidade para função f(x), USANDO O MÉTODO DA ACEITAÇÃO-REJEIÇÃO')


axe1.plot(x, y, label='f(x)', lw=2)
axe1.plot(x, uniform.pdf(x), color='green', label='g(x)', lw=2)
axe1.plot(x, M*uniform.pdf(x), color='red', label='Mg(x)', lw=2)
axe1.legend()
axe1.grid()
axe1.set_xlabel('x')
axe1.set_ylabel('y')
axe1.set_title('Comparação f(x), g(x), Mg(x)')

plt.show()
```

```python

def fx(x, lam):
    X = []
    for a in x:
        if(a < 0):
            X.append(0)
        else:
            X.append(lam*np.exp(-lam*a))
    return X

def inversa_acumuladaExp(u, lam):
    return - np.log(1 - u)/lam
```

```python

num = 10000
u = uniform.rvs(size=num)
amostra0, amostra1 = inversa_acumuladaExp(u, 2), inversa_acumuladaExp(u, 10)

x = np.linspace(0,max(max(amostra0), max(amostra1)),num)
py0, py1 = 2,10
y0, y1 = fx(x, py0), fx(x, py1)


fig, (axe1, axe2) = plt.subplots(1,2, figsize=(15,5))

axe1.hist(amostra0, density=True, edgecolor='black', label='Dados gerados para exp({0})'.format(py0))
axe1.plot(x, y0, lw=2, color='red', label='Função densidade de probabilidade para exp({0})'.format(py0))
axe1.legend()
axe1.set_xlabel('x')
axe1.set_ylabel('Frequência/desidade de probabilidade')
axe1.set_title('Histograma/Função densidade de probabilidade distribuição exp({0}),\ndados gerados com o método da TRANSFORMADA INVERSA '.format(py0))

axe2.hist(amostra1, density=True, edgecolor='black', label='Dados gerados para exp({0})'.format(py1))
axe2.plot(x, y1, lw=2, color='red', label='Função densidade de probabilidade para exp({0})'.format(py1))
axe2.legend()
axe2.set_xlabel('x')
axe2.set_ylabel('Frequência/desidade de probabilidade')
axe2.set_title('Histograma/Função densidade de probabilidade distribuição exp({0}),\ndados gerados com o método da TRANSFORMADA INVERSA '.format(py1))


plt.show()
```

```python

#definição g(x), distribuição exponencial, da qual sei amostrar
def gx(x, lam=1):
    X = np.array([])
    for a in x:
        if(a < 0):
            X = np.append(X,0)
        else:
            X = np.append(X,(lam*np.exp(-lam*a)))
    return X


#definição de fx, da qual quero amostrar
def fx(x):
    X = np.array([])
    for a in x:
        if(a < 0):
            X = np.append(X,0)
        else:
            X = np.append(X,(4.5*(a**2)*np.exp(-2*a)))
    return X

#definição fx/gx
def fxPorGx(x):
    X = np.array([])
    for a,b in zip(fx(x), gx(x)):
        if(a == 0 or b == 0):
            X = np.append(X, 0)
        else:
            X = np.append(X, a/b)
    return X
```

```python

num = 10000

# Para o ponto x = 2, temos f(x):
M = fxPorGx([2])

x = np.linspace(-4,4,num) #intervalo arbitrário
y = fx(x); 
yg = gx(x)

fig, axe = plt.subplots(1, figsize=(12,5))

plt.plot(x,y, label='f(x)', lw=2)
plt.plot(x, yg, label='g(x)', lw=2)
plt.plot(x, M*yg, label='Mg(x) = {0:.2f}g(x)'.format(*M), lw=2)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparação f(x), g(x) e Mg(x)')

plt.show()
```

Observe que acima M foi encontrado calculando-se f(x)/g(x) para x = 2, em que x = 2, foi encontrado achando-se 
o ponto de máxima da razão dessas funções, usando derivada. Uma outra alternativa para achar esse valor poderia ser, útil quando achar o ponto máximo usando derivada for díficil:


```python

M_2 = fxPorGx(x)
M_2 = max(M_2)
print('Valor de M calculado de outra maneira: {0:.2f}'.format(M_2))
```

```python

num = 10000
U = uniform.rvs(size=num, loc=0, scale=4) # distribuição da qual sei amostrar, usando a uniforme, [loc, loc + scale], ou seja gerando o intervalo [0,4]
quantValRegeitados = 0

X = []
for Y  in U:
    f = fx([Y])[0]
    g = M #(M*uniform(loc=0, scale=4).pdf(Y))[0] = M, pois g é uma distribuição uniforme, tendo pdf = 1, 
    r = f/g
    aceita = binom(1,r).rvs(size=1)
    if(aceita == 1):
        X.append(Y)
    else:
        quantValRegeitados +=1


fig, axe = plt.subplots(figsize=(12,4))

axe.hist(X, density=True, edgecolor='black', label='Dados gerados')
axe.plot(x, y, lw=2, color='red', label='Função densidade de probabilidade alvo')
axe.legend()
axe.set_xlabel('x')
axe.set_ylabel('Frequência/distribuição dados')
axe.set_title('Dados gerados usando o método da ACEITAÇÃO REGEIÇÃO')

plt.show()

print('Quantidade de valores usados: {0}\nQuantidade de valores aceitos: {1}\nQuantidade de valores rejeitados: {2}'.format(num, num-quantValRegeitados, quantValRegeitados))
```

<b>É expressamente importante e necessário que o suporte de g(x), a função da qual sei amostrar, seja igual ou maior que o suporte de f(x), a função da qual quero amostrar.</b>

```python

# definindo as constantes
B = 1.02e-4
c0 = 1.0855

cLinha = B/np.log(c0)

x = np.linspace(0,100,200)

# definindo a função de densidade de probabilidade usando a scipy.stats.gompertz
pdf_gompertz = []
for a in x:
    pdf_gompertz.append(np.log(c0)*gompertz.pdf(a, c= cLinha, loc=a*(1-np.log(c0))))

#os parãmetros, c e loc e a multiplicação por np.log é feita para que a distribuição de gompertz gerada
# tenha a mesma forma de f(x), a distribuição apresentada na questão, confira [5].


# definindo a função de densidade de probabilidade manualmente, observe que é equivalente a usar 
# a biblioteca scipy.stats definindo os parâmetros acima.
def pdf_gompertz_manual(x):
    B = 1.02e-4
    c0 = 1.0855
    return B*(c0**x)*np.exp(-(B/np.log(c0))*((c0**x)-1))


# PLOTS

fig, (axe1, axe2) = plt.subplots(1,2, figsize=(16,4))

axe1.plot(x, pdf_gompertz, lw=2)
axe1.set_title('Função densidade de probabilidade usando a biblioteca scipy.stats.gompertz')
axe1.set_xlabel('x (tempo de vida de adultos, anos)')
axe1.set_ylabel('probabilidade')
axe1.set_xticks(np.arange(0,101, 10))

axe2.plot(x, pdf_gompertz_manual(x), lw=2)
axe2.set_title('Função densidade de probabilidade definida manualmente')
axe2.set_xlabel('x (tempo de vida de adultos, anos)')
axe2.set_ylabel('probabilidade')
axe2.set_xticks(np.arange(0,101, 10))

plt.show()
```

- Utilizando as plotagens acima a estimativa para P(X < 40) = 0.10
- P(X ∈ [k, k+1]) parece ter um máximo próximo de 80.
- Díficil estimar isso pois essa curva serve melhor para uma estimativa para adultos, mas a probabilidade de morrer com mais de 100 anos aparenta ser maior que a te morrer antes dos 10.

A inversa da função acumulada é mostrada abaixo:

<img src='imgs/q7Inversa.jpeg' style='width:350px'>


```python

def inversaAcumuladaFx(u):
    B = 1.02e-4
    c0 = 1.0855
    return np.log(1 - (np.log(c0)*np.log(1-u))/B)/np.log(c0)
```

```python

num = 50000
U = uniform.rvs(size=num)

amostraGerada = inversaAcumuladaFx(U)


fig, axe = plt.subplots(figsize=(10,4))
axe.hist(amostraGerada, density=True,edgecolor='black', label='Dados gerados')
axe.plot(x, pdf_gompertz_manual(x), lw=2 , color='red', label='Função densidade de probabilidade alvo')
axe.set_title('Dados gerados para a distribuição alvo com o método da TRANSFORMADA INVERSA')
axe.set_xlabel('x (tempo de vida de adultos, anos)')
axe.set_ylabel('frequência/densidade de probabilidade')
axe.set_xticks(np.arange(0,101, 10))
axe.legend()

plt.show()
```

A probabilidade condicional P(X > 80|X > 50) é calculada abaixo:

```python

def probXMaior50(amostra):
    maior50 = 0
    for a in amostra:
        if(a > 50):
            maior50 += 1
    return maior50/len(amostra)


def probXMaior80IntersecaoMaior50(amostra):
    maior80 = 0
    for a in amostra:
        if(a > 80 and a > 50): #equivale a > 80
            maior80 += 1
    return maior80/len(amostra)

probXMaior80dadoXMaior50 = probXMaior80IntersecaoMaior50(amostraGerada)/probXMaior50(amostraGerada)

print('P(X > 80|X > 50) = {0}'.format(probXMaior80dadoXMaior50))
```

```python

#função calculo do rendimento
def h(t):
    return 2*np.exp(0.05*t)

def inversa_h(u):
    return np.log(u/2)/0.05

x = np.linspace(0,100,200)

fig, axe = plt.subplots(figsize=(8,4))
axe.plot(x, h(x), lw=2, color='red')
axe.set_xlabel('t (tempo de vida em anos)')
axe.set_ylabel('Dinheiro gerado (rendido) mil')
axe.set_title('Curva rendimento dado pela função h(t) = 2exp(0.05t)')

plt.show()
```

Calculando a inversa da função h, rendimento, para a entrada de 100, ou seja, buscando a idade para qual tem-se um rendimento de 100.000 mil, obtemos o resultado abaixo:

```python

t = inversa_h(100); print('Idade mínima para a empresa de seguros não ter prejuizo no seguro oferecido: ',t)
```

Arredondando o valor acima, temos que a idade mínima, que um cliente precisa viver, para que a empresa não sofra prejuízo ao oferecer o seguro é de 79 anos.<br>
Para os 50.000 clientes para o qual realizamos o experimento de monte carlo, num = 50000, temos, para a perda e ganho estimados:

```python

def perdaGanhoEstimada(amostra):
    perdaEstimada = 0
    ganhoEstimado = 0
    menor79 = 0
    for a in amostra:
        if(a < 79):
            perdaEstimada += h(a) - 100 # o que foi gerado, ganho menos os 100 que deverão ser pagos
            menor79 += 1
        else:
            ganhoEstimado += h(a) - 100
    return perdaEstimada, ganhoEstimado, menor79/len(amostra)

perdaE, ganhoE, probPerda = perdaGanhoEstimada(amostraGerada)
print('Perda estimada: {0:.2f}\nGanho estimado: {1:.2f}\n\nProbabilidade perda: {2:.2f} '.format(perdaE, ganhoE, probPerda))
```

Podemos observar que o ganho estimado é maior, em módulo, que a perda estimada. Observe também que a probabilidade de perda corresponde a probabilidade de uma pessoa morrer antes dos 79 anos, oferecendo assim uma perda para a empresa de seguros, mas que não significa prejuízo ao olharmos a perda e ganho total para uma amostra de 50 mil clientes gerados por monte carlo seguindo a distribuição de Gompertz.
