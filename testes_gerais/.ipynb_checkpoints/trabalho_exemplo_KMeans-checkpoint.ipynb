## Objetivo

Mostrar de maneira rápida o fucionamento do <b>KMeans</b>, algoritmo de clusterização, e uma aplicação utilizando Python.

```python

#libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from sklearn import cluster
```

### Iris Dataset

<img src='./iris.png'>

```python
iris = pd.read_csv('./Iris.csv', usecols=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm','PetalWidthCm','Species'])
```

```python
iris.head()
```

```python
iris.info()
```

```python
iris.describe().T
```

```python
sns.pairplot(iris, hue='Species')
plt.show()
```

```python
iris2D = iris[['SepalLengthCm', 'SepalWidthCm']]
```

```python
iris2D
```

```python
fig, axe = plt.subplots(figsize=(8,5))
axe = sns.scatterplot(data=iris2D, x='SepalLengthCm', y='SepalWidthCm')
plt.show()
```

### KMeans

```python

kmeans = cluster.KMeans(3)
clusters = kmeans.fit_predict(iris2D)
```

```python
pd.value_counts(clusters)
```

```python
x_center, y_center = kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1]
```

```python

fig, (axe1, axe2) = plt.subplots(1,2, figsize=(14,5))

sns.scatterplot(data=iris2D, x='SepalLengthCm', y='SepalWidthCm', hue=clusters, ax=axe1)
axe1.scatter(x=x_center, y=y_center, s = 80, color='red')
sns.scatterplot(data=iris, x='SepalLengthCm', y='SepalWidthCm', hue='Species', ax=axe2)
plt.show()
```
