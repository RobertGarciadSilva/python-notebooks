### (d) Abaixo é realizado a plotagem para a posição da corda em t = 0 e e o formato da corda em t = 1, usando a linguagem python.

```python

# importanto bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python

#definindo a função f(x)
def fx(x, periodo = 6):
    if(x >= 0 and x <3):
        return (4/3)*x
    if(x >= 3 and x <= 6):
        return (-4/3)*x + 8
    

#definindo a função u(x,t)
def u_xt(x, t=1.0):
    m1, m2 = x + t, x - t
    result  = (fx(m1) + fx(m2))/2
    return result
```

```python

# plotanto a posição da corda em t = 0

x = np.linspace(0,6,100)
y = []
for a in x:
    y.append(fx(a))

plt.plot(x,y, color='red', lw=3)
plt.title('Posição da corda em t = 0', size=15)
plt.show()
```

```python

# Plotando o formato da corda em t = 1

x_n = np.linspace(1,5,100)

y2 = []
for a in x_n:
    y2.append(u_xt(a))
    
plt.plot(x, y2, lw=3)
plt.title('Formato da corda em t = 1', size=15)
plt.show()
```
