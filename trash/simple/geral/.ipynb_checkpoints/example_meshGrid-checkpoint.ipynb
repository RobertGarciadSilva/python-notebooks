```python

x_1 = np.array([[1,2,3],[4,5,6]])
```

```python

print('x_1: ', x_1, end='\n\n')
print('x_1.T', x_1.T, end='\n\n')
print('x_1.ravel():\n',x_1.ravel(), end='\n\n')
print('x_1.T.ravel():\n',x_1.T.ravel(), end='\n\n')
```

```python

x_1 = np.array([[1,2,3],[4,5,6]])
x_2 = np.array([[7,8,9],[10,11,12]])

x_3 = [x_1.ravel(),x_2.ravel()]
x_4 = np.vstack(x_3)
x_5 = x_4.T

print('x_3:\n',x_3, end='\n\n')
print('x_4:\n',x_4, end='\n\n')
print('x_5:\n',x_5, end='\n\n')
```

```python

x = np.linspace(0,5,5)
y = np.linspace(0,5,5)

xGrid, yGrid = np.meshgrid(x,y)

xy = np.vstack([xGrid.ravel(), yGrid.ravel()]).T

# z = f(x,y) = sin(x^2 + y^2)
z = np.sin(xy[:,0]**2 + xy[:,1]**2) 

plt.contour(xGrid,yGrid, z.reshape(xGrid.shape))
plt.show()

```

```python

from sklearn.datasets import make_blobs

X, y = make_blobs(100, 2, centers=2, random_state=42, cluster_std=2.7)

plt.scatter(X[:,0], X[:,1], c=y)
```

```python

r = np.exp(-(X**2).sum(1))
```

```python

from mpl_toolkits import mplot3d
from ipywidgets import interact, fixed

def plot_3D(elev=30, azim=30, X=X, y=y):
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:,0],X[:,1], r, c=y)
    ax.view_init(elev=elev, azim=azim)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('r')

interact(plot_3D, elev=[-90, 90], azip=(-180, 180),X=fixed(X), y=fixed(y));
```
